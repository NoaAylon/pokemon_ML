# Fire Vs Water

```python
fire=pokemons[(pokemons['Type 1']=='Fire')] #fire contains all fire pokemons
water=pokemons[(pokemons['Type 1']=='Water')]  #all water pokemons
grass=pokemons[(pokemons['Type 1']=='Grass')]  #all grass pokemons
plt.scatter(fire.Attack.head(50),fire.Defense.head(50),color='R',label='Fire',marker="*",s=50) #scatter plot
plt.scatter(water.Attack.head(50),water.Defense.head(50),color='B',label="Water",s=25)
plt.scatter(grass.Attack.head(50),grass.Defense.head(50),color='G',label="Grass",marker="^",s=25)
plt.xlabel("Attack")
plt.ylabel("Defense")
plt.legend()
plt.plot()
fig=plt.gcf()  #get the current figure using .gcf()
fig.set_size_inches(12,6) #set the size for the figure
plt.show()
```

```python
plt.subplots(figsize = (15,5))
plt.title('Total by Type')
sns.boxplot(x = "Type 1", y = "Total",data = pokemons)
plt.show()
```

```python
sns.heatmap(
    pokemons.groupby('Type 1').median().loc[:, 'HP':'Speed'], 
    linewidths=1,
    cmap='RdYlBu_r',
);

```

```python

pokemons['Atk - Def'] = pokemons['Attack'] - pokemons['Defense']
pokemons['Sp.Atk - Sp.Def'] = pokemons['Sp. Atk'] - pokemons['Sp. Def']
no_legends = pokemons[pokemons.Legendary == False]
sns.heatmap(
    no_legends.groupby('Type 1').median().loc[:, 'HP':'Speed'], 
    linewidths=1,
    cmap='RdYlBu_r'
);
```

```python
sns.set_style("whitegrid")
with sns.color_palette([
    "#8ED752", "#F95643", "#53AFFE", "#C3D221", "#BBBDAF",
    "#AD5CA2", "#F8E64E", "#F0CA42", "#F9AEFE", "#A35449",
    "#FB61B4", "#CDBD72", "#7673DA", "#66EBFF", "#8B76FF",
    "#8E6856", "#C3C1D7", "#75A4F9"], n_colors=18, desat=.9):
    plt.figure(figsize=(12,10))
#     plt.ylim(200, 800)
    sns.swarmplot(x="Type 1", y="Total", data=no_legends, hue="Type 1", dodge=False, size=7)
    plt.legend(bbox_to_anchor=(1, 1), loc=1, borderaxespad=0.);
```

```python

sns.set(style="white", context="talk")
f, (ax1, ax2) = plt.subplots(2, 1, figsize=(15, 8), sharex=True)
data = no_legends.groupby("Type 1").mean()[['Speed', 'HP', 'Atk - Def', 'Sp.Atk - Sp.Def']]

sns.barplot(data.index, data['Atk - Def'], palette="RdBu_r", ax=ax1)
ax1.set_ylabel("Diverging in Atk - Def")

sns.barplot(data.index, data['Sp.Atk - Sp.Def'], palette="RdBu_r", ax=ax2)
ax2.set_ylabel("Diverging in Sp.Atk - Sp.Def")
```
